# MVP

## Train versus Vehicle Emissions
***
### Question/Need:

The goal of this project is to allow climate-conscious individuals to compare the c02 emissions of passenger train travel versus passenger vehicle travel in the United States so they can make more informed decisions about their means of travel.

The data for this project primarily comes from the <a href="https://www.climatiq.io/">Climatiq API</a>.

The below interactive web app allows individuals to first select an origin city, then a route in which the origin city is located, then a destination city in the selected route and finally, view the total c02 emissions per passenger of this trip.
